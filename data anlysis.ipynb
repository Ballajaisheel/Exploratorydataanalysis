**Descriptive** **statistics**

category values and their min and max


```python
cat=df.pivot_table(index=['category'],aggfunc='size')
category=cat.to_frame()
category
```

```python
max_cat=category.max()
max_cat
#fruit
```

```python
min_cat=category.min()
min_cat
#spices and herbs
```

customertype and their min and max

```python
custype=df.pivot_table(index=['customer_type'],aggfunc='size')
customertype=custype.to_frame()
customertype
```

```python
cust_max=customertype.max()
cust_max
#non-member
```

```python
cust_min=customertype.min()
cust_min
#gold
```

paytype and their min and max

```python
patype=df.pivot_table(index=['payment_type'],aggfunc='size')
paytype=patype.to_frame()
paytype
```

```python
pay_max=paytype.max()
pay_max
#cash
```

```python
pay_min=paytype.min()
pay_min
#debit card
```

more buyers where quantity is greater than 4 and expenses is more than 5

```python

most_users=df[['category','customer_type','quantity','total']]

most_user_type=most_users[most_users['quantity'] >=4 ]
most_user_types=most_user_type[most_user_type['total']>5]
most_user_types

```
